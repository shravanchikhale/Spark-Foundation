# To find out Outliers

```python
sns.boxplot(data["Sales"])
```

```python
sorted(data["Sales"])
```

```python
quantile1, quantile3= np.percentile(data["Sales"],[25,75])
```

```python
print(quantile1, quantile3)
```

```python
#To find IQR
iqr = quantile3 - quantile1
print(iqr)
```

```python
lower_bound = quantile1 -(1.5 * iqr)
upper_bound = quantile3 +(1.5 * iqr)
```

```python
data.drop(data[data["Sales"]> upper_bound].index, inplace = True)
data.drop(data[data["Sales"]< lower_bound].index, inplace = True)
```

```python
quantile1, quantile3= np.percentile(data["Profit"],[25,75])
iqr = quantile3 - quantile1
lower_bound = quantile1 -(1.5 * iqr)
upper_bound = quantile3 +(1.5 * iqr)
```

```python
data.drop(data[data["Profit"]> upper_bound].index, inplace = True)
data.drop(data[data["Profit"]< lower_bound].index, inplace = True)
```
